```python
dfavg = df.groupby('name', sort=False)[['quote.USD.percent_change_1h', 'quote.USD.percent_change_24h','quote.USD.percent_change_7d', 'quote.USD.percent_change_30d']].mean()
```

```python
dfstack = dfavg.stack()
```

```python
dfframe = dfstack.to_frame(name='values')
dfframe
```

```python
num = len(dfframe)

index = pd.Index(range(num))
dfindexed = dfframe.set_index(index)
dfindexed
```

```python
num = len(dfframe)

index = pd.Index(range(num))
dfindexed = dfframe.reset_index()
dfindexed
```

```python
dfindexed = dfindexed.rename(columns={'level_1': 'percent_change'})
```

```python
dfindexed['percent_change'] = dfindexed['percent_change'].replace(['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d'],['1h','24h','7d','30d'])
```
